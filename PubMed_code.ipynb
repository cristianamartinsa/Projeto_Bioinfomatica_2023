# PubMed

```python
from Bio import Entrez

keywords = input("Keywords for search:")

def search(query):
    handle = Entrez.esearch(db='pubmed',
                            sort='relevance',
                            retmax='20',
                            retmode='xml',
                            term=query)
    results = Entrez.read(handle)
    return results

def fetch_details(id_list):
    ids = ','.join(id_list)
    handle = Entrez.efetch(db='pubmed',
                           retmode='xml',
                           id=ids)
    results = Entrez.read(handle)
    return results

results = search(f'{keywords}')
id_list = results['IdList']
papers = fetch_details(id_list)
for i, paper in enumerate(papers['PubmedArticle']): print("%d) %s" % (i+1, paper['MedlineCitation']['Article']['ArticleTitle']))
```

This code returns a json with the chosen article

```python
import json
import pandas as pd
n = int(input("Which article do you want to read?"))
res = json.dumps(papers['PubmedArticle'][n-1], indent=2, separators=(',', ':'))
print(res)
```

This code opens an url with the article

```python
urls = []
for pubmed_id in id_list:
    url = f"https://pubmed.ncbi.nlm.nih.gov/{pubmed_id}/"
    urls.append(url)
```

```python
import webbrowser

n = int(input("Enter the number of the article to view its abstract:"))

url = urls[n-1]
webbrowser.open(url)
```

# PMC

```python
from Bio import Entrez
user_email = input("User e-mail:")
Entrez.email = user_email
keywords = input("Keywords for search:")
retmax = 10 

handle = Entrez.esearch(db="pmc", sort="relevance", term=keywords, retmax=retmax)
record = Entrez.read(handle)
id_list = record["IdList"]

titles = []
urls = []
for pmc_id in id_list:
    handle = Entrez.esummary(db="pmc", id=pmc_id)
    record = Entrez.read(handle)
    title = record[0]["Title"]
    url = f"https://www.ncbi.nlm.nih.gov/pmc/articles/{pmc_id}"
    titles.append(title)
    urls.append(url)

for i, title in enumerate(titles):
    print(f"{i+1}. {title}")
```

```python
import webbrowser
n = int(input("Enter the number of the article to view the full-text:"))

chosen_id = id_list[n-1]
handle = Entrez.efetch(db="pmc", id=chosen_id, retmode="xml")
record = Entrez.read(handle)

url = urls[n-1]
webbrowser.open(url)
```
